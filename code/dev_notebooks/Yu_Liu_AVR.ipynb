# Coding up the Yu and Liu Age-Velocity dispersion relation (AVR)

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Load the data for metal rich stars, metal poor stars, and all stars.

```python
mp = pd.read_csv("../../data/metal_poor_avr.csv")
mr = pd.read_csv("../../data/metal_rich_avr.csv")
a = pd.read_csv("../../data/all_stars_avr.csv")
```

```python
plt.errorbar(mp.Age_Gyr, mp.sigma_z_kms, xerr=mp.age_err, yerr=mp.sigma_z_err, fmt=".")
plt.errorbar(mr.Age_Gyr, mr.sigma_z_kms, xerr=mr.age_err, yerr=mr.sigma_z_err, fmt=".")
plt.errorbar(a.Age_Gyr, a.sigma_z_kms, xerr=a.age_err, yerr=a.sigma_z_err, fmt=".")
plt.xlabel("Age [Gyr]")
plt.ylabel("sigma z [km/s]")
```

Fit with v as x and age as y in log, but then plot age as x and v as y.

```python
import aviary as av
```

```python
p_mp, p_mr, p_a = av.get_avr_coefficients()

plt.errorbar(mr.Age_Gyr, mr.sigma_z_kms, xerr=mr.age_err, yerr=mr.sigma_z_err, fmt=".")
plt.xlabel("Age [Gyr]")
plt.ylabel("sigma z [km/s]")

vs = np.linspace(8, 30, 100)
ages = av.v_to_age(vs, p_mr)
plt.plot(ages, vs)

vs2 = av.age_to_v(ages, p_mr)
plt.plot(ages, vs2)
```
